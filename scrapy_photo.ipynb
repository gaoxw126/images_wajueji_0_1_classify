```python
#引入包
import requests
import time
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
from lxml import html
```

```python
# 设置列表页面URL的固定部分
url = 'https://product.cehome.com/wajueji/'
```

```python
# 最好在http请求中设置一个头部信息，否则很容易被封ip
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
'Accept':'text/html;q=0.9,*/*;q=0.8',
'Accept-Charset':'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
'Accept-Encoding':'gzip',
'Connection':'close',
'Referer':'http://www.baidu.com/link?url=_andhfsjjjKRgEWkj7i9cFmYYGsisrnm2A-TN3XZDQXxvGsM9k9ZZSnikW2Yds4s&amp;amp;wd=&amp;amp;eqid=c3435a7d00146bd600000003582bfd1f'
}
```

```python
#定义获取总页数的页面
def get_allpages(res,pattern):
    all_pages = res.find_all('div', 'digg')
    all_page = all_pages[0].find_all('span')[-1].text
    match = pattern.search(all_page).group()
    print("总页数长度" + str(match))
    return int(match)
```

```python
# 循环抓取列表页信息
def  get_pages_res(all_pages,target,BASE_URL_U2):
    for i in range(1,all_pages + 1):  #分页
        if i == 1:
            i=str(1)
            var_url = (target + '/' + i + BASE_URL_U2)
            r = requests.get(url=var_url, headers=headers)
            html = r.content

        else:
            i=str(i)
            #print(i)
            var_url=(target + '/' + i + BASE_URL_U2)
            var_url=requests.get(url=var_url,headers=headers)
            html2=r.content
            html = html + html2
            # 每次间隔1秒
            time.sleep(5) 
        # 解析抓取的页面内容
        res_pic = BeautifulSoup(html, 'html.parser') 
    return res_pic
```

```python
#获取所有图片的url
def get_images(res_pic,imgre):
    all_pictures = res_pic.find_all('div', 'searchResultBox')
    pictures = [ x.find_all('img')[0]   for x in all_pictures]
    images = [ imgre.search(str(x)).group().strip()[5:-1] for x in pictures]
    return images
```

```python
def save_images(root,target,images_list):
    path = root + target.split("/")[-1]
    print(path)
    try:
        if not os.path.exists(path):
            os.mkdir(path)
        for x in range(len(images_list)):
            r = requests.get(images_list[x])
            r.raise_for_status()
            #使用with语句可以不用自己手动关闭已经打开的文件流
            file = path + "//" + str(x) + ".png"
            print(file)
            with open(file,"wb") as f: #开始写文件，wb代表写二进制文件
                f.write(r.content)
            print("文件保存完成")     
    except Exception as e:
        print("文件保存失败:"+str(e))
```

```python
#抓取列表
base_urls = ['caterpillar','case','kubota','doosan','komatsu','sany','volvo','xcmg']
BASE_URL_U2 = '_15/ss_0_0/'
```

```python
for base_url in base_urls:
    target = url + base_url
    #抓取页面
    r = requests.get(url=target, headers=headers)
    html = r.content
    #BeautifulSoup 解析页面
    res = BeautifulSoup(html, 'html.parser')
    pattern = re.compile('[1-9]\d*')
    #获得总页数
    all_pages = get_allpages(res,pattern)
    print(all_pages)
    #正则图片链接
    res_pic = get_pages_res(all_pages,target,BASE_URL_U2)
    imgre = re.compile(r'src="(.*?)"')
    #得到所有图片链接list
    images =  get_images(res_pic,imgre)
    root = "D://wajueji//"
    save_images(root,target,images)
```
